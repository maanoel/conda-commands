```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread("pessoas.jpg")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagem)
```

```python
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(imagem_gray)
```

```python
classificador = cv2.CascadeClassifier("classificadores/hearcascade_frontalface_default.xml")
```

```python
faces = classificador.detectMultiScale(imagem_gray, 1.3,5) 
# O segundo parmetro define a reduo da imagem que ser 30% at o terceiro  a rigidez da anlise, quando maior, 
#mais criteriosa ser a anlise, fazendo o algoritimo de Haar demorar mais
```
