```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread("imagem01.jpg"); # abre uma imagem com opencv
```

```python
plt.imshow(imagem) #A imagem aberta é em RGB
```

```python
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagem_rgb)
```

```python
imagem_rgb.shape
```

```python
imagem_gray = cv2.cvtColor(imagem_rgb, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(imagem_gray, cmap="gray")
```

```python
imagem_gray.shape
```

```python
imagem = cv2.imread('pessoas.jpg')
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagem)
```

```python
imagem.shape
```

```python
imagem_roi = imagem[200:600, 0:450]
```

```python
plt.imshow(imagem_roi)
```

```python
cv2.imwrite("imagem_roi.png", imagem_roi)
```

```python
imagem_roi_bgr = cv2.cvtColor(imagem_roi, cv2.COLOR_RGB2BGR)
```

```python
cv2.imwrite("imagem_roi.png", imagem_roi_bgr)
```
